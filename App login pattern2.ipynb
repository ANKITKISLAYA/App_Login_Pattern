## calculating intervals from datetime

```python
for i in range(len(c)) :
    l = (c.iloc[i, 4]) 
    for j in range(len(l)-1):
        tstamp1 = l[j]
        tstamp2 = l[j+1]
        if tstamp1 > tstamp2:
            td = tstamp1 - tstamp2
        else:
            td = tstamp2 - tstamp1
        l[j] = td.days
    l1 = l[0:len(l)-1]
    if (len(l1)!=0):
        avg = sum(l1)/(len(l1))
        c.iloc[i,2] = avg
        c.iloc[i,3] = len(l)
    else :
        c.iloc[i,2] = 0
        c.iloc[i,3] = len(l)
```
